```python

import numpy as np
import matplotlib.pyplot as plt


def gaussian_2D(mean,covariance,num_points=100):
  return np.random.multivariate_normal(mean, covariance, num_points)
```

```python
'''
a) mean = [0,0], cov = [[1,0],[0,1]]
'''

mean = np.asarray([0,0])

covariance = np.asarray([[1,0],
                         [0,1]])
num_points = 100

points = gaussian_2D(mean,covariance,num_points)

x1 = points[:,0]
x2 = points[:,1]

plt.scatter(x1,x2)
plt.title('Gaussian 2D distribution, mean = [0,0], cov = [[1,0],[0,1]]')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

```python
'''
b) Change the mean to [-1,-1]
'''

mean = np.asarray([-1,-1])

covariance = np.asarray([[1,0],
                         [0,1]])
num_points = 100

points = gaussian_2D(mean,covariance,num_points)

x1 = points[:,0]
x2 = points[:,1]

plt.scatter(x1,x2)
plt.title('Gaussian 2D distribution, mean = [-1,-1], cov = [[1,0],[0,1]]')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

```python
'''
c) Double the variance of each component
'''

mean = np.asarray([0,0])

covariance = np.asarray([[2,0],
                         [0,2]])
num_points = 100

points = gaussian_2D(mean,covariance,num_points)

x1 = points[:,0]
x2 = points[:,1]

plt.scatter(x1,x2)
plt.title('Gaussian 2D distribution, mean = [0,0], cov = [[2,0],[0,2]]')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```
